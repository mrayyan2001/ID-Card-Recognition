```python
img = cv2.imread("./imgs/id3.png")
gray_img = Image.grayscale(img)

# Threshold
_, thr = cv2.threshold(gray_img, 145, 255, cv2.THRESH_BINARY)

# Noise reduction using gaussian and median
gaussian = cv2.GaussianBlur(thr, (3, 3), 0)
median = cv2.medianBlur(gaussian, 3)



adaptive = cv2.adaptiveThreshold(
    median, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 3, 2
)


Image.show(img, gray_img)
Image.show(gaussian, median)
Image.show(thr, adaptive)
```
